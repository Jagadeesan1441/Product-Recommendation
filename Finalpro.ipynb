```python
#function for scrolling a page till end of the page
def scroll_and_collect():
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
```

```python
def page(page_num):
    try:
        pagex_path=f'//a[text()="{page_num}"]'
        page_button=driver.find_element(By.XPATH,pagex_path)
        page_button.click()
        time.sleep(6)
    except Exception as e:
        print(f"erros occurs {e}")
        return False
    return True
```

```python
import pandas as pd
import re

# Function to clean reviews
def clean_review(review):
    """
    Cleans a review by removing unwanted characters, symbols, and emojis.
    """
    # Remove URLs
    review = re.sub(r"http\S+|www\S+|https\S+", '', str(review), flags=re.MULTILINE)

    # Remove emojis
    review = re.sub(r'[\U0001F600-\U0001F64F'  # emoticons
                    r'\U0001F300-\U0001F5FF'  # symbols & pictographs
                    r'\U0001F680-\U0001F6FF'  # transport & map symbols
                    r'\U0001F1E0-\U0001F1FF'  # flags (iOS)
                    r'\U00002700-\U000027BF'  # Dingbats
                    r'\U000024C2-\U0001F251]', '', review)

    # Remove special characters (except alphanumeric and basic punctuation)
    review = re.sub(r"[^a-zA-Z0-9.,!?'\s]", '', review)

    # Remove extra whitespaces
    review = re.sub(r'\s+', ' ', review).strip()

    return review

# Apply the clean_review function to the 'review' column
if 'review' in df1.columns:
    df1['cleaned_review'] = df1['review'].apply(clean_review)
else:
    print("The 'review' column is not found in df1.")

# Display the DataFrame with cleaned reviews
print(df1[['review', 'cleaned_review']].head())
```

```python
import nltk
nltk.download("vader_lexicon")
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid=SentimentIntensityAnalyzer()
df1["scores"]=df1["cleaned_review"].apply(lambda review:sid.polarity_scores(review))
df1.head(2)
```

```python
df1["compound"]=df1["scores"].apply(lambda d:d["compound"])
```

```python
def compound(score):
  if score>=0.05:
    return "positive"
  elif score<=-0.05:
    return "negative"
  else:
    return "neutral"
  
df1["vs_sentiment"]=df1["compound"].apply(compound)
```

```python
def create_product_advice(entry):
    product = entry['product_name']
    sentiment_score = entry['compound']
    review_excerpt = entry['review'][:60]  # Adjusted the character limit for the review excerpt
    
    if sentiment_score >= 0.75:  # Slightly adjusted threshold for high sentiment
        return f"Highly recommend {product}! Customers rave about its {review_excerpt}..."
    elif sentiment_score >= 0.5:  # Slightly adjusted threshold for moderate sentiment
        return f"Consider {product} for a reliable option. Customers mention: {review_excerpt}..."
    else:
        return f"You might want to check out other options before picking {product}. Some reviews note: {review_excerpt}..."

# Apply the advice function to the DataFrame
df1['advice'] = df1.apply(create_product_advice, axis=1)
```
